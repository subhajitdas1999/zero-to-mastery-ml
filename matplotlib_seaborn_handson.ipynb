```python
import pandas as pd

df = pd.read_csv("data/covid-data.csv",delimiter=';')
```

```python
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df['Date_reported'] = pd.to_datetime(df['Date_reported'], format='%d/%m/%Y')
```

```python
df_country = df[df['Country'] == 'Italy']
plt.figure(figsize=(12, 6))
plt.plot(df_country['Date_reported'], df_country['Cumulative_cases'], marker='o', linestyle='-', color='r', label='Cumulative Cases')
plt.xlabel('Date')
plt.ylabel('Cumulative Cases')
plt.title('Cumulative COVID-19 Cases in Italy')
plt.legend()
plt.grid(True)
plt.show()
```
